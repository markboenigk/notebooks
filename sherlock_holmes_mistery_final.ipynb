```python
####################################### ################
#Author: Felipe Saraiva Moreira de Sena / MBAN-DD / 2023
########################################################
"""
Created on Sun Jan 29 1:42:00 2023
@author: Felipe Saraiva Moreira de Sena
"""
"""
    DocString:
    
    A) Introduction:
    
    Welcome to the Sherlock Holmes Mystery Game!
    
    In this game, you will take on the role of Sherlock Holmes, the world's 
    greatest detective. You have been called to investigate a mysterious crime
    that has occurred at 100 Education Street.
    Your task is to gather clues and evidence, and use your reasoning skills to
    solve the crime.
    
    The game has some clues, all of them are multiple-choice question.
    
    The game's objective is to solve the crime by correctly identifying the 
    criminal, based on the clues provided.
    
    Game version: 1.0
       
    Stage 1: Investigate the crime
    Stage 2: Interrogate suspects
    Stage 3: Confront the suspect and solve the crime
    
    B) Known Bugs and/or Errors:
    None.
    
"""
##############################################################################
######################### import random all packages used in the game#########
##############################################################################
import random
import time
from sys import exit
import random

##############################################################################
############################ CODE: SHERLOCK HOLMES MISTERY GAME ##############
################################# Auxiliar functions: ########################
##############################################################################

#function to keep asking user for input until a valid answer is provided or 
#max_tries is reached
def valid_input(prompt='', max_tries=3, valid_answers = ['1', '2', '3','yes','YES', 'y', 'no', 'NO', 'n']):
    """
    This function prompts the user for input and keeps asking until a valid 
    answer is provided.
    A valid answer can be one of the following:
    - Yes or no questions: 'yes', 'YES', 'y', 'no', 'NO', 'n'
    - Questions that require input of 1, 2 or 3: '1', '2', '3'
    If the user fails to provide a valid answer after max_tries attempts, 
    they will be asked if they want to try again.
    If the input is a valid affirmative answer it will return True, 
    otherwise False.
    """
    user_input = ''
    tries = 0

    # Keep asking user for input until a valid answer is provided or max_tries
    #is reached
    while True:
        user_input = input(prompt).lower()
        # Check if input is in the valid_answers list
        if user_input in valid_answers:
            break
        tries += 1
        # If max_tries is reached, ask user if they want to try again
        #Nested Conditional Statements are used here
        if tries == max_tries:
            print("You have failed to provide a valid answer after", max_tries, "tries.")
            user_input = input("Last Chance. Do you want to try again? (yes/no)")
            if user_input in ['yes', 'YES', 'y']:
                return valid_input(prompt)
            elif user_input in ['no', 'NO', 'n']:
                print("You have chosen not to try again. Goodbye.")
                break
            else:
                print("You have failed to provide a valid answer. Goodbye.")
                break

    # Return input if input is a valid numeric answer
    if user_input in ['1', '2', '3']:
        return user_input
    # Return True if input is a valid affirmative answer, False otherwise
    else:
        return user_input in ['yes', 'YES', 'y']

# #Define a function called thinking and print "Thinking...Thinking...Thinking"
#to improve the user experience.

def think():
    num_seconds = 3
    for countdown in reversed(range(num_seconds + 1)):
        if countdown > 0:
            print("Thinking", end='...')
            time.sleep(1)
            
def list_student_names():
    names = []
    surnames = ["Smith", "Johnson", "Williams", "Jones", "Brown", "Davis", \
                "Miller", "Wilson", "Moore", "Taylor", "Anderson", "Thomas",\
                "Jackson", "White", "Harris", "Martin", "Thompson", "Garcia",\
                "Martinez", "Robinson", "Clark", "Rodriguez", "Lewis", "Lee",\
                "Walker", "Hall", "Allen", "Young", "Hernandez", "King",\
                "Wright", "Lopez", "Hill", "Scott", "Green", "Adams", "Baker",\
                "Gonzalez", "Nelson", "Carter", "Mitchell", "Perez", "Roberts",\
                "Turner", "Phillips", "Campbell", "Parker", "Evans", "Edwards",\
                "Stewart", "Flores", "Morris", "Nguyen", "Murphy", "Rivera",\
                "Cook", "Rogers", "Morgan", "Peterson", "Cooper", "Reed", \
                "Bailey", "Bell", "Gomez", "Kelly", "Howard", "Ward", "Cox",\
                "Diaz", "Richardson", "Wood", "Watson", "Brooks", "Bennett",\
                "Gray", "James", "Reyes", "Cruz", "Hughes", "Price", "Myers",\
                "Long", "Foster", "Sanders", "Ross", "Morales", "Powell",\
                "Sullivan", "Russell", "Ortiz", "Jenkins", "Gutierrez", \
                "Perry", "Butler", "Barnes", "Fisher"]

    required_names = ["Jonathan Brown", "Joana Jones", "John Bridgerton"]
    for i in range(70):
        if i < 3:
            names.append(required_names[i])
        else:
            name = random.choice(surnames)
            surname = random.choice(surnames)
            names.append(name + " " + surname)
    random.shuffle(names)
    return names

# Function that tells the user have solved the crime
def case_solved():
    print("""\nCorrect! Congrats, you solved the mystery! 
\n\t- The initials in the pen drive "JB" means "Jonathan Brown" in Clue 1
\t- "Jonathan Brown" is a men as described by the witness in Clue 2
\t- Finally "Jonathan Brown" has 13 characters, the final piece from Clue 3

You are a great detective. Sherlock is proud of you!
""")
    exit('The End.')
    
# Function that tells the user have faild to solve the crime
def fail():
    print("""\nYou have failed in solve the crime. Game Over!""")
    exit('Game Over!')

##############################################################################
############################ Stage 1: Investigate the crime ##################
##############################################################################
 
# Ask the user if want to play:
def start_game():
    lets_play = valid_input(prompt="""Would you like to play a game to solve a crime?
\n\t- Type (Y) for YES
\t- Type (N) for NO \n""", max_tries=3, valid_answers = ['yes', 'YES', 'y', 'no', 'NO', 'n'])

    # Check if the player's answer is correct
    if lets_play == True:
        next
    else: 
        exit("You decided to exit the game.")

def stage_1():
    think()
    print(f"""\n
\tWelcome to the Sherlock Holmes Mystery Game!

\tYou are Detective {user_name}, auxiliar of the world's greatest detective Sherlock Holmes.
\tYou have been called to investigate a crime at Holmes Business School.
\tA student is suspected of stealing and hacking the computer of the school 
\tto change his grades in the system.

\tProfessor Watson was reviewing students grades, and noticed that something was wrong.
\tThere was no A grades for that cohort. Somehow, 3 students got A's their transcripts.

\tYour task is to gather clues and evidence, and use your reasoning skills to solve the crime.
    
\tThis game has three stages:
    
\tStage 1: Investigate the crime
\tStage 2: Interrogate suspects
\tStage 3: Confront the suspect and solve the crime
    
\tGood luck on your journey detective {user_name}!
    """)
    
    input("<To continue press 'Enter' (Stage 1)>\n")

    think()
    
    # First stage: Gather initial clues and evidence

    ###########################
    #### Clue 1: USB drive ####
    ###########################

    print("""\n
Clue 1: You find a USB drive near the computer that shouldn't be there. The inicials 'JB' were printed in the pen-drive.
\n\t- 1. A presentation
\t- 2. A movie
\t- 3. Hack codes
""")
    
    clue_1 = valid_input(prompt="What type of data do you think is stored on the USB drive?",\
                         max_tries=3, valid_answers = ['1', '2', '3'])

    #Convert the input in integer
    clue_1 = int(clue_1)

    # Check if the player's answer is correct
    if clue_1 == 3:
        print("\nCorrect! The USB drive contains hack codes.")
    else:
        print("\nIncorrect. The USB drive contains hack codes.")    
    
    input("<To continue press 'Enter'>\n")
    
    think()

    ##################################
    #### Clue 2: Witness statement ###
    ##################################
    
    while True:
        print("""\n
Clue 2: The witness reported observing an individual walking in the school's 
hallway during class hours. The person, who couldn't be identified by face, 
was described as having short hair, wearing a hoodie, and carrying a laptop bag.
\n\t- 1. The suspect is a man.
\t- 2. The suspect is a woman.
""") 
        clue_2 = input()
        try:
            clue_2 = int(clue_2)
            if clue_2 in [1,2]:
                if clue_2 == 1:
                    print("\nCorrect! The suspect is a man.")
                    break
                else:
                    print("\nIncorrect. The suspect is a man.")
                    break
            else:
                print("Invalid input, please try again.")
        except:
            print("Invalid input, please try again.")

###########################################################################################################
# NOTE: The code below would be more efective, since I have a function to check the validity of the input #
#However for the sake of practice/exercise, I used Try and Except to showcase a second possible way       #
#to check user inputs. See commented below an alternative for the same result using valid_input().        #
#                                                                                                         #
#     print("""\n                                                                                         #
# Clue 2: The witness reported observing an individual walking in the school's                            #
# hallway during class hours. The person, who couldn't be identified by face,                             #
# was described as having short hair, wearing a hoodie, and carrying a laptop bag.                        #
# \n\t- 1. The suspect is a man.                                                                          #
# \t- 2. The suspect is a woman.                                                                          #
# """)                                                                                                    #
#     clue_2 = valid_input(prompt="Do you think the suspect is a man or woman?",\                         #
#                          max_tries=3, valid_answers = ['1', '2'])                                       #
#                                                                                                         #
#     #Convert the input in integer                                                                       #
#     clue_2 = int(clue_2)                                                                                #
#                                                                                                         #
#     # Check if the player's answer is correct                                                           #
#     if clue_2 == 1:                                                                                     #
#         print("\nCorrect! The suspect is a man.")                                                       #
#     else:                                                                                               #
#         print("\nIncorrect. The suspect is a man.")                                                     #
###########################################################################################################

    input("<To continue press 'Enter'>\n")
    
    think()
    
    #################################
    #### Clue 3: Computer Records ###
    #################################
    
    print("""\n
Clue 3: You examine the school's computer records and find that an unknow ID have alterated the grades. 

System traces show that the student's full name is 13 letters long, but unfortunately only the first letter
of the name was revealed. The hacker name starts with the letter "J".

What would you like to do?
\n\t- 1. Give up the case. It is impossible to find out who commited the crime.
\t- 2. See the list of all student names.
\t- 3. See the list of student names that start with "J"
""")    
    clue_3_action = valid_input(prompt="Insert a number of your choice:",\
                         max_tries=3, valid_answers = ['1', '2', '3'])

    #Convert the input in integer
    clue_3_action = int(clue_3_action)

    # Check if the player's answer is correct
    if clue_3_action == 3:
        think()
        print("""\nSmart move! There are three students with the name starting with 'J' 
\n\t- "John Bridgerton", "Joana Jones", and "Jonathan Brown".

Congrats! You gathered enough evidence to know who is guilty.
You will jump directly to stage 3: Confront the suspect!
""")        
        input("<To continue press 'Enter'>\n")
        think()
        stage_3()
    elif clue_3_action == 2:
        print("\nHere is the list of all student names:")
        print(list_student_names())
        print("""\n
Congrats! You are close to solve this mistery.
Now you have the chance to interview the suspects (Stage 2).
""")    
        input("<To continue press 'Enter'>\n")
        think()
        stage_2()
    else:
        fail()

##############################################################################
############################ Stage 2: Interrogate suspects ###################
##############################################################################

def stage_2():
    #Second stage: Interview one of the suspects    
    print("""\n
Step 2: After long hours of interviews, there are only 3 left to be interviewed. 
We're running out of time. if we don't identify the hacker within the next hour, 
per school policy, grades cannot be challenged. 

You only have time to interview one suspect.

Who would you like to interview?
\n\t- 1. "John Bridgerton" 
\t- 2. "Joana Jones"
\t- 3. "Jonathan Brown"
""") 
    
    interview_stage2 = valid_input(prompt="Insert a number of your choice:",\
                         max_tries=3, valid_answers = ['1', '2', '3'])
    
    #Convert the input in integer
    interview_stage2 = int(interview_stage2)

    # Check if the player's answer is correct
    if interview_stage2 == 1:
        # Dialogue suspect == "John"
        print(f"""
\n\tJohn: Hey, what's up? You here to question me about the grade tampering case?
\tDetective {user_name}: Yes, that's right. Can you tell me more about the grades you received?
\tJohn: Yeah, I got an A in the class, but it's not like I cheated or anything.
\tI just had a really good teacher.
\tDetective {user_name}: I see. Can you tell me more about your teacher?
\tJohn: He's a great guy, really helpful. Sometimes he's change grades when he
\tnotice that there was an issue with the quiz. I thought I got an A after his review.
\tWas not that what happened?
\tDetective {user_name}: I see. Thank you for your time, John. You will know more soon. I have to go now.
""")
        stage_3()
    elif interview_stage2 == 2:
        # Dialogue suspect == "Joana"
        print(f"""
\n\tJoana: Hello, Detective Holmes. What can I do for you today?
\tDetective {user_name}: I am investigating a case of grade tampering at Homes Business School. 
\tI understand you were one of the students who received an A grade. Can you tell me more about it?
\tJoana: Oh, that. Yes, I did receive an A grade. I studied very hard for the final exam and I deserved it.
\tDetective {user_name}: I see. However you didn't have an A before. How do you explain the change?"
\tJoana: Professor Watson always review our grandes when there are more than one correct answer
\tI believe he updated my grade from the last quiz.
\tDetective {user_name}: I see. Thank you for your time, Joana. I have to go now.
""")
        stage_3()
    else:
        # Dialogue suspect == "Jonathan"
        print(f"""
\n\tJonathan: Hi. Is something the matter?
\tDetective {user_name}: Yes, I'm here to ask you about the grade tampering case.
\tJonathan: Oh, that. I-I don't know anything about that.
\tDetective {user_name}: Then how do you explain receiving an A in the class?
\tJonathan: I-I-I don't know, maybe it was a mistake or something.
\tDetective {user_name}: A mistake? Are you sure you didn't have any help?
\tJonathan: N-no, I swear. I didn't do anything wrong.
\tDetective {user_name}: Your behavior and responses are raising suspicions, Michael.
\tJonathan: Look, I-I can't tell you anything else. I-I need a lawyer.
\tDetective {user_name}: Very well, we'll continue this conversation later.
""")
        stage_3()

##############################################################################
################ Stage 3: Confront the suspect and solve the crime ###########
##############################################################################

def stage_3():
    #Third stage: Confront the suspect    
    print("""
\nStage 3: Who would you like to acuse of the crime?
\n\t- 1. "John Bridgerton" 
\t- 2. "Joana Jones"
\t- 3. "Jonathan Brown"
""") 
    
    confront_stage3 = valid_input(prompt="Insert a number of your choice:",\
                         max_tries=3, valid_answers = ['1', '2', '3'])
    
    #Convert the input in integer
    confront_stage3 = int(confront_stage3)

    # Check if the player's answer is correct
    if confront_stage3 == 3:
        case_solved()
        
    elif confront_stage3 == 2:
        print("""\nIncorrect! You acused the wrong person.
\nAfter Sherlock Holmes have a look at the case evidence he rapidly found out the truth:
\n\t- The initials in the pen drive "JB" means "Jonathan Brown" not "Joana Jones" in Clue 1
\t- "Joana Jones" is a woman. The witness saw a men in Clue 2 
\t- Finally "Joana Jones" has 10 characters, the final piece from Clue 3 
\tsays the hacker's name was 13 characters long

All evidences showed that your guess was wrong. You're Fired!
""")
        fail()
    else:
        print("""\nIncorrect! You acused the wrong person.
\nAfter Sherlock Holmes have a look at the case evidence he rapidly found out the truth:
\n\t- The initials in the pen drive "JB" means "Jonathan Brown" not 
\t"John Bridgerton" in Clue 1 (This one was hard)
\t- "John Bridgerton" is a men. The witness saw a men in Clue 2 (This one was hard too)
\t- Finally "John Bridgerton" has 14 characters, the final piece from Clue 3 
\tsays the hacker's name was 13 characters long. It was clear that it couldn't be "John Bridgerton".

\tThere was one evidences against your guess. Youre Fired!
""")
        fail()
        
##############################################################################
##################################### Starting the game: #####################
##############################################################################

# Starting the game!
start_game()

# I added user_name here because we were not allowed to use global variables, 
#and I want to use the user_name variable during the game.
user_name = input("Awesome! Let's have some fun, but before we start. What is your name?\n").title()
input("<To continue press 'Enter'>\n")
stage_1()
```
